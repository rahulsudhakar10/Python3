Step 1:Import random modules.This will use shuffle deck .Declare variables to store suits ranks.

```python
import random
suits =('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen',
        'King','Ace')
values ={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,
        'Queen':10,'King':10,'Ace':11}

playing = True
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return self.rank +" of "+self.suit
    
```

Step 3:Creat a deck class we have 52 cards and shuffle that latter

```python
class Deck():
    
    def __init__(self):
        self.deck=[]
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def __str__(self):
        deck_comp=''
        for card in self.deck:
             deck_comp += '\n'+ card.__str__()
            
        return "The deck has"+deck_comp
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card=self.deck.pop()
        return single_card
        
        
```

Step 5 :Create a hand class holding card object fetched from deck

```python
class Hand:
    
    def __init__(self):
        self.cards=[]
        self.value= 0
        self.aces= 0
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank] 
        
        if card.suit == 'ACE':
            self.aces=1
        
        
    def adjust_for_ace(self):
        
        #when value is greater tha 21 Then my ace should be changed to 
        #1 intsead of 11
        while self.value > 21 and self.aces > 0:
            self.value -= 10
            self.aces  -= 1
        
    
```

```python
class Chip:
    
    def __init__(self,total=100):
        self.total= total
        self.bet=0
        
    def win_bet(self):
        self.total += self.bet 
    
    def loose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet= int(input('How many chips do want to place'))
        except TypeError:
            print('Please enter a valid number')
        else:
            if chips.bet > chips.total:
                print(f'U dont have enough chips total chips available are {chips.total}')
            else:
                break
```

```python
#Taking a card from deck is a hit
def hit(deck,hand):
    single_card=deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    global playing
    
    while True:
        x= input('Hit or stand? enter h for hit and stand for s')
        
        if x == 'h':
            hit(deck,hand)
        elif x == 's':
            print('player stands dealer turn!')
        else:
            print("please enter 'h' or 's'")
            continue
        break    
```
